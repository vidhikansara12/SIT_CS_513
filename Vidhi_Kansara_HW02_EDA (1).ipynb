```python
# VK_20029476
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

1. Load the dataset in python

```python
# VK_20029476
df = pd.read_csv("breast-cancer-wisconsin.csv")
df.head()
```

I. Summarizing each column (e.g. min, max, mean)

```python
# VK_20029476
print("Summary statistics of each column:")
print(df.describe())
```

II. Identifying missing values

```python
# VK_20029476
print("Missing values in each column:")
print(df.isnull().sum())
```

III. Replacing the missing values with the “mean” of the column

```python
# VK_20029476
df.replace('?', np.nan, inplace=True)
df['F6'] = pd.to_numeric(df['F6'])
df.fillna(df.mean(), inplace=True)
print("Missing values after replacing with mean:")
print(df.isnull().sum())
```

IV. Displaying the frequency table of “Class” vs. F6

```python
# VK_20029476

# 'Class' column is the target column and 'F6' is a feature
frequency_table = pd.crosstab(df['Class'], df['F6'])
print(frequency_table)
```

V. Displaying the scatter plot of F1 to F6, one pair at a time

```python
# VK_20029476

features = ['F1', 'F2', 'F3', 'F4', 'F5', 'F6']
# Creating a pairplot with enhanced features
sns.pairplot(df[features], diag_kind='kde', plot_kws={'alpha':0.7, 's':30}, palette='viridis')
# Adding a title to the pairplot
plt.suptitle("Pairwise Relationships between Features F1 to F6", size=16, y=1.02)
# Display the plot
plt.show()
```

VI. Show histogram box plot for columns F7 to F9

```python
# VK_20029476

sns.boxplot(data=df[['F7', 'F8', 'F9']], palette='Set2')
# title
plt.title("Box Plot for Features F7, F8, F9", size=16)
# Adding axis labels
plt.xlabel("Features", size=12)
plt.ylabel("Values", size=12)
# Display the plot
plt.show()
```
